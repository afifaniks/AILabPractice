# Linear Regression w/ SKLearn

```python
# Importing Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
import seaborn as si
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading data
df = pd.read_csv('datasets/Weather.csv', sep=',')

df.head()
```

```python
df.shape
```

```python
df.plot(x='MinTemp', y='MaxTemp', style='o')
plt.title('MinTemp vs MaxTemp')
plt.xlabel('MinTemp')
plt.ylabel('MaxTemp')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
si.distplot(df['MaxTemp'])
```

```python
X = df['MinTemp'].values.reshape(-1, 1)
Y = df['MaxTemp'].values.reshape(-1, 1)
```

```python
# Splitting data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Model
regression_model = LinearRegression()

regression_model.fit(x_train, y_train)
```

```python
print("Intercept:", regression_model.intercept_)
print("Slope:", regression_model.coef_)
```

```python
# Prediction
predict = regression_model.predict(x_test)

df_predict = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': predict.flatten()})

df_predict
```

```python
# Plotting Result
plt.scatter(x_test, y_test)
plt.plot(x_test, predict, color='red', linewidth=2)
plt.show()
```
